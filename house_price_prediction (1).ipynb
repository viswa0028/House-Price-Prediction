# **Data preprocessing**

Filling all the necessary values with mode or mean or median .

```python
df['MSZoning'] = df['MSZoning'].fillna(df['MSZoning'].mode())
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())
```

```python
df['Utilities'] = df['Utilities'].fillna(df['Utilities'].mode())
```

```python
df['Exterior1st'] = df['Exterior1st'].fillna(df['Exterior1st'].mode())
df['Exterior2d'] = df['Exterior2nd'].fillna(df['Exterior2nd'].mode())
```

```python
df['MasVnrType'] = df['MasVnrType'].fillna(df['MasVnrType'].mode()[0])
df['MasVnrArea'] = df['MasVnrArea'].fillna(df['MasVnrArea'].mean())
df['BsmtCond'] = df['BsmtCond'].fillna(df['BsmtCond'].mode()[0])
df['BsmtQual'] = df['BsmtQual'].fillna(df['BsmtQual'].mode()[0])
df['BsmtExposure'] = df['BsmtExposure'].fillna(df['BsmtExposure'].mode()[0])
df['BsmtFinType1'] = df['BsmtFinType1'].fillna(df['BsmtFinType1'].mode()[0])
df['BsmtFinType2'] = df['BsmtFinType2'].fillna(df['BsmtFinType2'].mode()[0])
df['GarageType'] = df['GarageType'].fillna(df['GarageType'].mode())
df['GarageFinish'] = df['GarageFinish'].fillna(df['GarageFinish'].mode()[0])
df['GarageQual'] = df['GarageQual'].fillna(df['GarageQual'].mode()[0])
df['GarageCond'] = df['GarageCond'].fillna(df['GarageCond'].mode()[0])
```

```python
df.drop(['Fence','PoolQC','MiscFeature','FireplaceQu'],axis = 1,inplace = True)
```

```python
df.dropna()
```

```python
df.shape
```

**Creating dummies**

```python
listed = ['MSZoning','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood',
         'Condition2','BldgType','Condition1','HouseStyle','SaleType','SaleCondition','ExterCond',
         'ExterQual','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
         'RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','Heating','HeatingQC',
         'CentralAir','Electrical','KitchenQual','Functional','GarageType','GarageFinish','GarageQual',
         'GarageCond','PavedDrive'] #these are all the things that needs to be having dummies
```

```python
for i in listed:
  df = pd.concat([df,pd.get_dummies(df[i])],axis = 1)
  df = df.drop(i,axis = 1) #creating dummies and dropping the main column
```

Removing all the duplicate values

```python
df = df.loc[:,~df.columns.duplicated()].copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Exterior2d'] = le.fit_transform(df['Exterior2d']) #This is one of the column that needs to transformed so imported
```

```python
from sklearn.model_selection import train_test_split as tst
```

```python
X_train,X_test,y_train,y_test = tst(df.drop(['SalePrice'],axis =1),df.SalePrice,test_size = 0.2)
```

# Training the model xgboost

```python
import xgboost
model = xgboost.XGBRegressor()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

The accuracy came out to be 90.65% but if we use some other models then it might be higher or lower.
